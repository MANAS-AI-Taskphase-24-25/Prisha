# standardizing function

```python
def standardize_features(X):
    mean = np.mean(X, axis = 0)
    X_std = ( X - mean )/X.std()
    std = np.std(X, axis = 0)
    std[std==0] = 1
    return X_std, std, mean

X_standardized, mean, std = standardize_features(X)
```

```python
# generating polynomial features
```

```python
def generate_polynomial_features(X, degree=2):
    poly_features = X.copy()
    for d in range(2, degree + 1):
        for i in range(X.shape[1]):
            poly_features = np.column_stack((poly_features, X[:, i] ** d))
    return poly_features
```

```python
# training linear regression model
```

```python
def linear_regression(X, y):
    X_intercept = np.column_stack((np.ones(X.shape[0]), X))
    theta = np.linalg.inv(X_intercept.T.dot(X_intercept)).dot(X_intercept.T).dot(y)
    return theta
```

```python
# making predictions
```

```python
def predict(X, theta):
    X_intercept = np.column_stack((np.ones(X.shape[0]), X))
    predictions = X_intercept.dot(theta)
    return predictions
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def r_squared(y_true, y_pred):
    total_variance = np.sum((y_true - np.mean(y_true)) ** 2)
    explained_variance = np.sum((y_true - y_pred) ** 2)
    return 1 - (explained_variance / total_variance)
```

```python
# applying the functions
```

```python
X_poly = generate_polynomial_features(X_standardized, degree=2)
```

```python
theta = linear_regression(X_poly, y)
```

```python
y_pred = predict(X_poly, theta)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r_squared(y, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R):", r2)
```
